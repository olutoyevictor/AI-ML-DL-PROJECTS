## Activation functions

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Testing on linear data

```python
def linear_data(X):
    return 2*X + 1
```

```python
X = np.arange(0, 51)
y = linear_data(X)
```

```python
X
```

```python
y
```

```python
plt.scatter(X, y)
plt.show()
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
    loss='mse',
    metrics=['mse']
)
```

```python
model.fit(X, y, epochs=800)
```

```python
model.get_weights()
```

```python
y_pred = model.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

PERFECT PREDICTIONS!

### 2. Testing on non-linear data

```python
def non_linear(X):
    return 2*X**2 + 1
```

```python
X = np.arange(0, 51)
y = non_linear(X)
```

```python
X
```

```python
y
```

```python
plt.scatter(X, y)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64),
    tf.keras.layers.Dense(64),
    tf.keras.layers.Dense(1)
])
```

```python
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mse']
)
```

```python
model.fit(X, y, epochs=3500)
```

```python
y_pred = model.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

AWFUL!

### Testing on non-linear data with activation functions

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
    loss='mse',
    metrics=['mse']
)
```

```python
history = model.fit(X, y, epochs=2000)
```

```python
y_pred = model.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

We nailed it!
